# Prueba de una hipótesis estadística

Una **hipótesis estadística** es una afirmación o conjetura sobre una o más poblaciones. La estructura de la prueba de hipótesis se formulará con el uso del término hipótesis nula y se denota por $H_0$.  El rechazo de $H_0$ conduce a la aceptación de una hipótesis alternativa, denotado $H_1$.  
Posibles situaciones para probar una hipótesis estadística:


| Tabla         | $H_0$ es verdadera    | $H_0$ es falsa        |
| :------------| :------------------- | :------------------- |
|No rechaza $H_0$   |Decisión correcta      |Error tipo II$=\beta$  |
|Rechaza $H_0$  |Error tipo I$=\alpha$  |Decisión correcta     |

**Error tipo I** es rechazar la hipótesis nula cuando es verdadera:No hay diferencia en los tratamientos. **Falso Positivo**.


**Error tipo II** es aceptar la hipótesis nula cuando es falsa: Existen diferencias en los tratamientos. **Falso Negativo**.

![image.png](tiposIyII.png)



Una prueba de alguna hipótesis estadística, donde la alternativa es unilateral, como

$H_0: \theta = \theta_0   $  $\hspace{2cm}$     ó quizás $\hspace{2cm}$          $   H_0: \theta = \theta_0$  

$H_1:\theta >\theta_0$   $\hspace{5.6cm}$              $H_1: \theta < \theta_0$  $\hspace{1cm}$  se llama **prueba de una sola cola**

Una prueba de alguna hipótesis estadística, donde la alternativa es biilateral, como

$H_0: \theta = \theta_0$     

$H_1: \theta \neq \theta_0$   $\hspace{1cm}$  se llama **prueba de dos colas**

Procedimiento para prueba de hipótesis con probabilidad de error fijo de tipo I.

1. Defina las hipótesis nulas y alternativas.
2. Elija un nivel de significancia fijo  .
3. Elija un estadístico adecuado y establezca la región crítica basada en el nivel de significancia. 
4. De la estadística de prueba calculada, rechazar H0 si la estadística está dentro de la región crítica. De 
    lo contrario, no rechazar.
5. Sacar conclusiones científicas o de ingeniería.

## Ejemplo:

Una empresa que fabrica focos afirma que un tipo particular de foco durará 850 horas en promedio con una desviación estándar de 50. Un grupo de protección al consumidor cree que el fabricante ha sobreestimado la vida útil de sus focos alrededor de 40 horas. ¿Cuántas focos tiene que probar el grupo de protección del consumidor para comprobar su aseveración con una confianza razonable?


Para entender el problema primero suponemos que queremos tomar una muestra para el experimento para trabajarlo con un error de tipo I de 0.1, una prueba de dos colas con las hipótesis

$$H_0:\mu = 85$$
$$H_0:\mu \neq 85$$

consideramos una muestra 

```r
m=c(82,83,87,76,79,88,76)
media<-mean(m)
sigma<-sd(m)
nd<-length(m)
media
sigma
nd
t<-(media-85)/(sigma/sqrt(nd))
print(paste("estadístico t = ",t))
min<- -qt(.95,nd-1)
max<-qt(.95,nd-1)
print("intervalo de aceptación de dos colas al 10% de significancia")
print(c(min,max))
```

```r
x<-seq(-8,8,.1)
t00<-dt(x,nd-1)
plot(x,t00,type='l',lwd=3,xlab='Valores t',col='blue',ylab='Probabilidad', main='Distribucion t')
legend("topleft","G.L.=7")
```

```r
N<-19
df<-N-1
x<-seq(-8,8,.1)

# Distribución t central
t0<-dt(x,df)
plot(x,t0,type='l',lwd=3,xlab='Valores t',col='blue',ylab='Probabilidad', main='Distribuciones t Centrales y No-Centrales')
legend("topright",c("ncp=0","ncp=3.48711"),lty=1,col=c('blue','green'),lwd=3)
legend("topleft","D.F.=18")
criticalt<-qt(.975,df)


# Distribución t no central

d<-(850-810)/50

ncp1<-sqrt(N)*d

t1<-dt(x,df,ncp1)
lines(x,t1,lwd=3,col='green')
criticalt1<-pt(criticalt,df,ncp1)
potencia<-1-criticalt1

print(paste("Tamaño del efecto d = ",d))
print(paste("Parámetro no central delta = ",ncp1))
print(paste("t crítico = ",criticalt))
print(paste("t crítico no central = ",criticalt1))
print(paste("beta = ",criticalt1))
print(paste("potencia = ",potencia))


```
